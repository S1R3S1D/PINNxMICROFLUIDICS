```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
input_path = "D:\\CAD\\pinn-geom-matlab\\"
data_file_int = "pinn-interior.asc"
point_cloud_int = np.loadtxt(input_path+data_file_int, skiprows=1)


data_file_bc_list = []
directory = os.fsencode(input_path)

for file in os.listdir(directory):
     filename = os.fsdecode(file)
     if filename != "pinn-interior.asc" and filename != "pinn-domain.ipynb":
        #point_cloud_bc = np.loadtxt(input_path+filename, skiprows=1)
        data_file_bc_list.append(filename)

point_cloud_inlet_1_d = np.loadtxt(input_path+data_file_bc_list[0], skiprows=1)
point_cloud_inlet_2_d = np.loadtxt(input_path+data_file_bc_list[1], skiprows=1)
point_cloud_inlet_c = np.loadtxt(input_path+data_file_bc_list[2], skiprows=1)
point_cloud_outlet = np.loadtxt(input_path+data_file_bc_list[3], skiprows=1)
point_cloud_walls = np.loadtxt(input_path+data_file_bc_list[4], skiprows=1)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(point_cloud_int[:,0], point_cloud_int[:,1], s=3)
plt.scatter(point_cloud_inlet_1_d[:,0], point_cloud_inlet_1_d[:,1], s=3)
plt.scatter(point_cloud_inlet_2_d[:,0], point_cloud_inlet_2_d[:,1], s=3)
plt.scatter(point_cloud_inlet_c[:,0], point_cloud_inlet_c[:,1], s=3)
plt.scatter(point_cloud_outlet[:,0], point_cloud_outlet[:,1], s=3)
plt.scatter(point_cloud_walls[:,0], point_cloud_walls[:,1], s=3)
plt.legend(["interior", "Inlet-1d", "Inlet-2d", "Inlet-c", "Outlet", "wall"], loc="best")
plt.show()
```
